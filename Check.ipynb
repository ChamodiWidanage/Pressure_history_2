```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Load data from an Excel file
# Replace 'your_file.xlsx' with the path to your Excel file
file_path1_o = Path("G:/Chamodi/LSDYNA3D/RP_TH_models/Check/2.75m18.5kg/1_original.xlsx")
file_path1_r = Path("G:/Chamodi/LSDYNA3D/RP_TH_models/Check/2.75m18.5kg/1_reduced.xlsx")
file_path2_o = Path("G:/Chamodi/LSDYNA3D/RP_TH_models/Check/2.75m18.5kg/2_original.xlsx")
file_path2_r = Path("G:/Chamodi/LSDYNA3D/RP_TH_models/Check/2.75m18.5kg/2_reduced.xlsx")
output_path = Path('G:/Chamodi/LSDYNA3D/RP_TH_models/Check/2.75m18.5kg/')

# Read the data into a DataFrame
df1_o = pd.read_excel(file_path1_o)
df1_r = pd.read_excel(file_path1_r)
df2_o = pd.read_excel(file_path2_o)
df2_r = pd.read_excel(file_path2_r)
```

```python
# Plot the data
plt.figure(figsize=(6, 6))
# Set global font to Times New Roman and font size
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.size'] = 20

# Example: Plotting columns named 'X' and 'Y'
# Replace 'X' and 'Y' with the actual column names from your Excel file
plt.plot(df1_o['Time'], df1_o['Pressure'], linestyle='-', color='cornflowerblue', linewidth=5, label='Original')
plt.plot(df1_r['Time'], df1_r['Pressure'], linestyle='--', color='r',linewidth=3, label='Reduced')

# Adding labels and title
plt.xlabel('Time (ms)')  # Replace with your x-axis label
plt.ylabel('Reflected pressure (kPa)')  # Replace with your y-axis label
plt.title('2.75 m - 18.5 kg - 0 degreees')  # Replace with your plot title

# Adding grid and legend
#plt.grid(True)
plt.legend()

plt.savefig(output_path/'0_degrees.png', dpi=200, bbox_inches='tight')
# Show the plot
plt.show()
```
