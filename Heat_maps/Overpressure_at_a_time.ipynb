```python
from pathlib import Path
import pandas as pd
```

```python
# define current work directory
current_dir = Path.cwd()

#Access input folder
input_folder = current_dir/"Angle_model/NumResults_14.6m27.1kg"

specific_time = 30
adj_time = 20
time_to_check = specific_time - adj_time
```

```python
# Make a list of data file names
files = list(input_folder.rglob("*"))

# Initialize a list to store results
results = []

# for loop for files
for file in files:
    df = pd.read_csv(file, delimiter="\s+", skiprows = 8, skipfooter = 1, header = None, names = ["time", "pressure"])
    
    # Find the index of the row where the time is closest to the specific time
    closest_index = (df['time']*1000 - time_to_check).abs().idxmin()
    
    # Get the corresponding pressure and file stem
    closest_time = df.loc[closest_index, 'time']*1000 + adj_time
    closest_pressure = df.loc[closest_index, 'pressure']/1000
    file_stem = file.stem
    
    # Append the result (file stem, closest time, and pressure) to the list
    results.append([file_stem, closest_time, closest_pressure])

# Create a new DataFrame to store the results
output_df = pd.DataFrame(results, columns=["file_name", "closest_time", "pressure"])

# Save the results to a CSV file
output_df.to_csv(current_dir/f"Angle_model/{specific_time}s_num_pressures.csv", index=False)
```
