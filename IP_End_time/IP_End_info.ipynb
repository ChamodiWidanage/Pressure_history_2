```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcdefaults()
```

```python
# Assuming your dataset is in a pandas DataFrame called 'df'
# You can replace 'df' with your dataset variable if it's different
df = pd.read_excel('End_time.xlsx')

# Drop the last column
df = df.drop(df.columns[-1], axis=1)

# Define the new column headers
new_headers = ['Explosive mass', 'Standoff distance', 'End time']  # Adjust as per the number of columns in your DataFrame

# Assign the new headers to the DataFrame
df.columns = new_headers

# Calculate the Spearman correlation matrix
spearman_corr = df.corr(method='spearman')

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(spearman_corr, dtype=bool), k=1)

# Set global font to Times New Roman (optional for aesthetics)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.size'] = 15

# Plot the heatmap using seaborn with the mask
plt.figure(figsize=(4, 3.5))
sns.heatmap(spearman_corr, mask=mask, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5,  vmin=-0.11, vmax=1)

# Add a title and labels
#plt.title('Spearman Correlation Coefficient Heatmap', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Show the heatmap
plt.tight_layout()
# Save the plot as an image
plt.savefig('IP_End_info.png', dpi=200, facecolor='white', bbox_inches='tight')
plt.show()
```
